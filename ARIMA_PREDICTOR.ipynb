```python
import pickle
from statsmodels.tsa.arima.model import ARIMA

# Function to train ARIMA model and make predictions
def train_arima_model(data, column_name):
    model = ARIMA(data[column_name], order=(5, 1, 0))  # Define ARIMA model order
    model_fit = model.fit()
    return model_fit

# Train models for each category
Regular_models = {col: train_arima_model(df, col) for col in gasoline_formulations_columns}
Premium_models = {col: train_arima_model(df, col) for col in conventional_gasoline_columns}
Reformulated_models = {col: train_arima_model(df, col) for col in reformulated_gasoline_columns}
Diesel_model = train_arima_model(df, 'D1')

# Function to save trained models using pickle
def save_models(models, filename):
    with open(filename, 'wb') as file:
        pickle.dump(models, file)

# Specify the filenames for saving the models
formulations_filename = 'formulations_models.pkl'
conventional_filename = 'conventional_models.pkl'
reformulated_filename = 'reformulated_models.pkl'
diesel_filename = 'diesel_model.pkl'

# Save formulations models
save_models(formulations_models, formulations_filename)

# Save conventional models
save_models(conventional_models, conventional_filename)

# Save reformulated models
save_models(reformulated_models, reformulated_filename)

# Save diesel model
with open(diesel_filename, 'wb') as file:
    pickle.dump(diesel_model, file)

print("Models saved successfully.")
```
